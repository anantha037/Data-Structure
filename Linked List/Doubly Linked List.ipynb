```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def array_to_linked_list(self, arr):
        for data in arr:
            self.add_at_end(data)
    
    def add_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        if self.head:
            self.head.prev = new_node
        self.head = new_node
    
    def add_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        new_node.prev = temp
    
    def delete_node(self, value):
        temp = self.head
        while temp and temp.data != value:
            temp = temp.next
        if not temp:
            return
        if temp.prev:
            temp.prev.next = temp.next
        if temp.next:
            temp.next.prev = temp.prev
        if temp == self.head:
            self.head = temp.next
        temp = None
    
    def delete_at_beginning(self):
        if self.head:
            self.head = self.head.next
            if self.head:
                self.head.prev = None
    
    def delete_at_end(self):
        if not self.head:
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        if temp.prev:
            temp.prev.next = None
        else:
            self.head = None
    
    def insert_after(self, x, data):
        temp = self.head
        while temp and temp.data != x:
            temp = temp.next
        if temp:
            new_node = Node(data)
            new_node.next = temp.next
            new_node.prev = temp
            if temp.next:
                temp.next.prev = new_node
            temp.next = new_node
    
    def insert_before(self, x, data):
        temp = self.head
        while temp and temp.data != x:
            temp = temp.next
        if temp:
            new_node = Node(data)
            new_node.next = temp
            new_node.prev = temp.prev
            if temp.prev:
                temp.prev.next = new_node
            else:
                self.head = new_node
            temp.prev = new_node
    
    def find_length(self):
        count = 0
        temp = self.head
        while temp:
            count += 1
            temp = temp.next
        return count
    
    def search(self, key):
        temp = self.head
        while temp:
            if temp.data == key:
                return True
            temp = temp.next
        return False
    
    def reverse(self):
        temp = None
        current = self.head
        while current:
            temp = current.prev
            current.prev = current.next
            current.next = temp
            current = current.prev
        if temp:
            self.head = temp.prev
    
    def print_list(self):
        temp = self.head
        while temp:
            print(temp.data, end=' <-> ')
            temp = temp.next
        print("None")
    
    def print_list_reverse(self):
        temp = self.head
        while temp and temp.next:
            temp = temp.next
        while temp:
            print(temp.data, end=' <-> ')
            temp = temp.prev
        print("None")
```

```python

# Example Usage
dll = DoublyLinkedList()
dll.array_to_linked_list([10, 20, 30, 40, 50])
dll.print_list()

dll.add_at_beginning(5)
dll.print_list()

dll.add_at_end(60)
dll.print_list()

dll.delete_node(30)
dll.print_list()

dll.reverse()
dll.print_list()

dll.print_list_reverse()
```
