```python
class Node:
    def __init__(self,key,value):
        self.key = key
        self.value = value
        self.next = None

class HashTable:
    def __init__(self,size):
        self.size = size
        self.table = [None for _ in range (size)]

    def _hash(self,key):
        return hash(key) % self.size

    def insert(self,key,value):
        idx = self._hash(key)
        head = self.table[idx]
        temp = head

        while temp:
            if temp.key == key:
                temp.value = value
                return 'Updated'
            temp  = temp.next
        new_node = Node(key,value)
        new_node.next = head
        self.table[idx] = new_node

    def delete(self,key):
        idx = self._hash(key)
        temp = self.table[idx]
        prev = None

        while temp:
            if temp.key == key:
                if prev:
                    prev.next = temp.next
                else:
                    self.table[idx] = None
                return 'Deleted'
            prev = temp
            temp = temp.next
        return 'Node not found'

    def get(self,key):
        idx = self._hash(key)
        temp = self.table[idx]

        while temp:
            if temp.key == key:
                return temp.value
            temp = temp.next
        return 'Not Found'

    def display(self):
        for i in range (self.size):
            temp = self.table[i]
            chain = []
            while temp:
                chain.append(f'({temp.key},{temp.value})')
                temp = temp.next
            print(f'{i} : {'-->'.join(chain) if chain else 'None'}')
```

```python
ht = HashTable(10)
ht.insert("apple", 5)
ht.insert("banana", 3)
ht.insert("grape", 10)
ht.insert("apple", 8)  # update value
ht.display()

print("Get 'banana':", ht.get("banana"))

print(ht.delete("banana"))
ht.display()

ht.delete("banana")
print("Get 'banana':", ht.get("banana"))
```
